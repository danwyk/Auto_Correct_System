## 2.1 Starts-with
The second task for today is to create a function that takes a string as input and returns all the words that start with that string as output. For example, entering `print(starts_with("ANTEP"))` should give us all the words that start with the letters `'ANTEP'`, including *antepenult*, *anteposition*, *antepast*, and a few others.

1. Name the function (something like `starts_with`), giving it one argument (labeled something like `letters`).
2. Within the function, create an empty dictionary `{}`. E.g. you might type `out = {}`.
3. Then, update `out` using a *for*-loop. The loop should iterate over the Scrabble dictionary called `dictionary` that we read in earlier (note: Python considers this object a <i>list</i>). For each word in the dictionary, check to see if the beginning of the word matches the function input `letters`. If it does, add the word to `out`. 
> WARNING: You might want to use "slice" notation like this: `if word[:len(letters)]==letters`. You will need to ensure that your function doesn't call for slices that don't exist; e.g. make sure the function doesn't ask for `word[:6]` for a word that's less than 7 letters long!
4. Finish the function with a `return` statement: `return out`.

````python
def starts_with(letters):
  words_list = []
  temp = ["antepenult", "anteposition", "antepast", "pastantepast", "an"]
  letters = letters.lower()
  max_length = len(letters)

  for word in temp:
    word = word.lower()

    if len(word) < max_length:
      max_length = len(word)

    if word[:max_length] == letters:
      words_list.append(word)
  
  return words_list

print(starts_with("ANTEP"))
````

## 2.2 Ends-with
Build the same kind of function, but for words that *end* with a given sequence of letters. For instance, `print(ends_with('TANGLE')` should return all the words that end with `TANGLE`, such as *detangle*, *rectangle*, and of course *tangle* itself, among a few other words.

````python
def ends_with(letters):
  words_list = []
  temp = ["detangle", "rectangle", "tangle", "tanglede"]
  letters = letters.lower()
  letters = letters[::-1]
  max_length = len(letters)
  # print(f"Letters: {letters}")

  for word in temp:
    word = word.lower()
    word = word[::-1]

    if len(word) < max_length:
      max_length = len(word)
    
    # print(f"Word: {word}")
    if word[:max_length] == letters:
      word = word[::-1]
      words_list.append(word)
  
  return words_list

print(ends_with("TANGLE"))
````
